### Data visualisation using Matplotlib

- plt.chartName()
- plt.xlabel()
- plt.ylabel()
- plt.title()
- plt.show()

```python
x1 = list(range(1,21))
y1 = np.power(x1,5)
y2 = np.log(x1)
y3 = np.exp(x1)
y4 = np.sqrt(x1)


print(f"x1 = {x1}")
print(f"y1 = {y1}")
print(f"y2 = {y2}")
print(f"y3 = {y3}")
print(f"y4 = {y4}")
```

```python
x1 = list(range(1,21))
y1 = np.power(x1,5)

plt.figure(figsize = (20,6))

plt.plot(x1,y1, c = 'red', marker = "d")
plt.xlabel("x1")
plt.ylabel("y1")
plt.title("x1 vs y1")
plt.show()
```

#### Subplot

- syntax : subplot(rows, cols, index)

```python
x1 = list(range(1,21))
y1 = np.power(x1,5)
y2 = np.log(x1)
y3 = np.exp(x1)
y4 = np.sqrt(x1)

plt.figure(figsize = (20,10))

plt.subplot(2,2,1)
plt.plot(x1,y1, c = 'green', marker = "d")
plt.xlabel("X1")
plt.ylabel("Y1")
plt.title("Chart: X1 vs Y1")

plt.subplot(2,2,2)
plt.plot(x1,y2, c = 'red', marker = "d")
plt.xlabel("X1")
plt.ylabel("Y2")
plt.title("Chart: X1 vs Y2")

plt.subplot(2,2,3)
plt.plot(x1,y3, c = 'blue', marker = "d")
plt.xlabel("X1")
plt.ylabel("Y3")
plt.title("Chart: X1 vs Y3")

plt.subplot(2,2,4)
plt.plot(x1,y4, c = 'yellow', marker = "d")
plt.xlabel("X1")
plt.ylabel("Y4")
plt.title("Chart: X1 vs Y4")


plt.show()
```

#### Data Anslysis & Chart Type:
- Univariate analysis - Explore one col at a time
    - Numerical - Histogram, Bar
    - Categorical - Count, Bar, Pie
- Bivariate analysis - Explore two col at a time
    - Numerical vs Numerical - Correlation, Line, Scatter
    - Categorical vs Numerical - Bar, Pie, Box
- Multivariate analysis - Explore more than two col at a time
    - (Numerical vs Numerical) vs Categorical - Correlation, Line, Scatter
    - (Categorical vs Numerical) vs Categorical - Bar, Pie, Box
    - All Numericals - Correlation

### Data Visualisation using Seaborn:

```python
import seaborn as sns  
```

#### Univariate analysis
- Numerical - Histogram, Bar

```python
plt.figure(figsize = (20,5))

sns.histplot(df['Price (INR)'], bins = 10, color = 'orange')
plt.xlabel("Price", fontsize = 14)
plt.ylabel("Distribution", fontsize = 14)
plt.title("Histogram for Price", fontsize = 18)
plt.show()
```

```python
df['Price (INR)'].describe()
```

#### Univariate analysis
- Categorical - Count, Bar, Pie

```python
plt.figure(figsize = (20,5))

sns.countplot(df['Gender'])
plt.xlabel("Gender", fontsize = 14)
plt.ylabel("Distribution", fontsize = 14)
plt.title("Count for Gender", fontsize = 18)
plt.show()
```

```python
plt.figure(figsize = (20,10))

sns.countplot(y = df['PrimaryColor'])
plt.xlabel("PrimaryColor", fontsize = 14)
plt.ylabel("Distribution", fontsize = 14)
plt.title("Count for PrimaryColor", fontsize = 18)
plt.show()
```

### Bivariate analysis 
- Numerical vs Numerical - Correlation, Line, Scatter

```python
plt.figure(figsize = (20,6))

sns.scatterplot(x = df['NumImages'], y = df['Price (INR)'], color = "magenta")
plt.xlabel("Number of Images", fontsize = 14)
plt.ylabel("Price Distribution", fontsize = 14)
plt.title("Price vs NumImages", fontsize = 18)
plt.show()
```

### Bivariate analysis 
- Categorical vs Numerical - Bar, Pie, Box

```python
plt.figure(figsize = (20,6))

sns.barplot(x = df['PrimaryColor'], y = df['Price (INR)'])
plt.xlabel("PrimaryColor", fontsize = 14)
plt.ylabel("Price Distribution", fontsize = 14)
plt.title("Price vs PrimaryColor", fontsize = 18)
plt.show()


# bar height = mean
# black vertical lines = std dev
```

```python
# Boxplot

plt.figure(figsize = (20,6))

sns.boxplot(x = df['Gender'], y = df['Price (INR)'])
plt.xlabel("Gender", fontsize = 14)
plt.ylabel("Price Distribution", fontsize = 14)
plt.title("Price vs Gender", fontsize = 18)
plt.show()

# Lower whisker, upper whisker, median, Q1(25%), Q3(75%)
# IQR = 75% - 25%
# lower Whisker = 25% - 1.5 * IQR
# Upper Whisker = 75% + 1.5 * IQR
```

```python
# stripplot

plt.figure(figsize = (20,6))

sns.stripplot(x = df['Gender'], y = df['Price (INR)'])
plt.xlabel("Gender", fontsize = 14)
plt.ylabel("Price Distribution", fontsize = 14)
plt.title("Price vs Gender", fontsize = 18)
plt.show()
```

#### Pandas Group By:

- Syntax: df.groupby(colName).agg(newColName = (oldColName, "Attribute")...)

```python
genderGroup = df.groupby('Gender').agg(totalRecords = ("Gender", 'size'),
                                        avgPrice = ('Price (INR)', 'mean'),
                                        avgImages = ('NumImages', 'mean'),
                                        maxPrice = ('Price (INR)', 'max'))
genderGroup                                         
```

```python
genderGroup.sort_values('totalRecords', ascending = False)
```

```python
genderGroup.index.values
```

```python
genderGroup.index.values[3]
```

```python
print(genderGroup['avgPrice'].values)

print(genderGroup.avgPrice.values)
```

```python
# Analysis of Gender group

plt.figure(figsize = (20, 8))
plt.suptitle("Analysis of Gender Group", fontsize = 20)

plt.subplot(1,4,1)
genderGroup = genderGroup.sort_values('totalRecords', ascending = False)
plt.bar(x = genderGroup.index.values, height = genderGroup.totalRecords.values,
        color = ['blue', 'orange', 'green', 'pink', 'yellow', 'magenta'])
plt.ylabel('Total Records')
plt.title('Total Records Distribution')
plt.xticks(rotation = 90)

plt.subplot(1,4,2)
genderGroup = genderGroup.sort_values('avgPrice', ascending = False)
plt.bar(x = genderGroup.index.values, height = genderGroup.avgPrice.values,
        color = ['blue', 'orange', 'green', 'pink', 'yellow', 'magenta'])
plt.ylabel('avgPrice')
plt.title('Avg Price Distribution')
plt.xticks(rotation = 90)

plt.subplot(1,4,3)
genderGroup = genderGroup.sort_values('avgImages', ascending = False)
plt.bar(x = genderGroup.index.values, height = genderGroup.avgImages.values,
        color = ['blue', 'orange', 'green', 'pink', 'yellow', 'magenta'])
plt.ylabel('avgImages')
plt.title('Avg Images')
plt.xticks(rotation = 90)


plt.subplot(1,4,4)
genderGroup = genderGroup.sort_values('maxPrice', ascending = False)
plt.bar(x = genderGroup.index.values, height = genderGroup.maxPrice.values,
        color = ['blue', 'orange', 'green', 'pink', 'yellow', 'magenta'])
plt.ylabel('maxPrice')
plt.title('Maximum Price')
plt.xticks(rotation = 90)

plt.show()
```

```python
genderGroup.index.values
```

```python
# pie chart

plt.figure(figsize = (10,8))

plt.pie(labels = genderGroup.index.values,
        x = genderGroup.avgPrice.values,
       shadow = True,
       explode = [0, 0.2, 0.3, 0, 0, 0],
       autopct = "%0.1f%%")

plt.show()
```
